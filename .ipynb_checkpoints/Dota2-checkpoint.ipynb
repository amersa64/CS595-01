```python
import json
#from pprint import pprint

with open('heroes.json') as data_file:    
    heroes = json.load(data_file)
```

```python
def heroMapping(data):
    heroRoles = ['carry', 'escape', 'nuker', 'durable','initiator','support','pusher','disabler','jungler']
    heroesAttr = ['attackRate','attackRange','projectileSpeed','movementSpeed','complexity']
    for role in heroRoles:
        data[role+str(1)]=np.zeros(data.shape[0])
        data[role+str(2)]=np.zeros(data.shape[0])
    for attr in heroesAttr:
        data[attr+str(1)]=np.zeros(data.shape[0])
        data[attr+str(2)]=np.zeros(data.shape[0])     
    i=0
    for row in data.iterrows():
        if(i % 100==0):
            print(i *100 / data.shape[0])
        for heroID in range(1,114):
            # If team 1
            #print(row)
            #print("Row: {}".format(row[1]))
            if(row[1][heroID]==1):
                for role in heroes[heroID-1]['roles']:
                    data.loc[i,str(role)+"1"]=data.loc[i,str(role)+"1"]+heroes[heroID-1]['roles'][str(role)]
                for attr in heroesAttr:
                    # Line of focus
                    data.loc[i,attr+"1"]=data.loc[i,attr+"1"]+heroes[heroID-1][attr]
            # If team 2
            if(row[1][heroID]==-1):
                for role in heroes[heroID-1]['roles']:
                    data.loc[i,str(role)+"2"]=data.loc[i,str(role)+"2"]+heroes[heroID-1]['roles'][str(role)]
                for attr in heroesAttr:
                    data.loc[i,attr+"2"]=data.loc[i,attr+"2"]+heroes[heroID-1][attr]
        i = i + 1
```

```python
data.drop(['attack1','attack2'], inplace=True, axis = 1)
```

```python
def runLR(data,datatest):
    LR = LogisticRegression()
    LR.fit(data.drop('win', axis = 1),data['win'])
    pred = LR.predict(datatest.drop('win',axis=1))
    print(accuracy_score(pred, datatest['win']))
```

```python
def runTree(data,datatest):
    clf = tree.DecisionTreeClassifier()
    clf.fit(data.drop('win', axis = 1),data['win'])
    pred_Tree = clf.predict(datatest.drop('win',axis=1))
    print(accuracy_score(pred_Tree, datatest['win']))
```
