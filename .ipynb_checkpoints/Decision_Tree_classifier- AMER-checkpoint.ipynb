```python
from sklearn import tree
import pandas
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import graphviz 
from sklearn import metrics
```

```python
file_name = 'train.csv'
```

```python
data = pandas.read_csv(file_name, index_col='PassengerId' )
```

```python
data.columns
```

```python
data.dropna(inplace=True)
```

```python
data.columns
```

```python
data.drop(['Name','Ticket', 'Cabin'],axis = 1,inplace = True)
```

```python
data['Embarked'] = LabelEncoder().fit_transform(data['Embarked'])
data['Sex'] = LabelEncoder().fit_transform(data['Sex'])
```

```python
train_x,test_x, train_y,test_y = train_test_split(data.drop('Survived', axis = 1), data['Survived'], test_size = 0.33)
```

```python
clf1 = tree.DecisionTreeClassifier(max_depth=1)
```

```python
clf1 = clf1.fit( train_x,train_y )
```

```python
pridict_d_1 = clf1.predict(test_x)
```

```python
print(metrics.accuracy_score(test_y, pridict_d_1))
```

```python
dot_data = tree.export_graphviz(clf1,feature_names=data.columns,filled=True, rounded=True,  special_characters=True,out_file=None) 
graph = graphviz.Source(dot_data) 
graph
```
